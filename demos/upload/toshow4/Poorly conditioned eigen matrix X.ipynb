```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 15
x = np.linspace(0,1,n)
X = np.zeros((n,n))
for i in range(n):
    X[:,i] = x**i
```

```python
print(X)
```

```python
for i in range(n):
    plt.plot(x, X[:,i])
```

```python
A = np.linalg.inv(X) @ np.diag(np.arange(1,n+1)) @ X
```

```python
np.linalg.cond(A)
```

```python
w, _ = np.linalg.eig(A)
print(w)
```
